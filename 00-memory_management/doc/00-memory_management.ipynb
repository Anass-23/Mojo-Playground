```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

SAVE_FIG: bool = False
DATA_PATH: Path = Path('./data/')
IMG_PATH:  Path = Path('./img/') 
```

```python
# CSV data
mojo_data = pd.read_csv(DATA_PATH / 'mojo.csv')
python_data = pd.read_csv(DATA_PATH / 'python.csv')

mojo_num_elements = mojo_data['num_elements']
mojo_time_ms = mojo_data['time_ms']

python_num_elements = python_data['num_elements']
python_time_ms = python_data['time_ms']
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(mojo_num_elements, mojo_time_ms, label='Mojo')
plt.plot(python_num_elements, python_time_ms, label='Python')

# X-axis ticks (500_000 step size) and X-axis to log10 scale
plt.xticks(range(0, max(mojo_num_elements.max(), python_num_elements.max()) + 1, 500_000))
plt.fill_between(mojo_num_elements, mojo_time_ms, python_time_ms, color='red', alpha=0.3, label='Time Difference')
plt.xscale('log')

# Titles...
plt.title('Performance Comparison of Summing Arrays (Ones) - Mojo vs Python')
plt.xlabel(r'Num. Elements ($log_{10}$)')
plt.ylabel(r'Time ($ms$)')
plt.legend()
plt.grid(True)

# Plot display and saving
if SAVE_FIG:
    plt.savefig(IMG_PATH / 'bench1.jpeg')
    plt.savefig(IMG_PATH / 'bench1.pgf')
plt.show()
```
